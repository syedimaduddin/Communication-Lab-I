### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal
```

### Formulae Used:
 Message signal -> $m(t)=A_ccos(2 \pi f_mt)$ <br>
 Carrier signal -> $c(t)=A_ccos(2 \pi f_ct)$ <br>
 AM signal -> $s(t)=A_c[1+K_a m(t)]\cos(2 \pi f_ct)$

```python
Am=1 # amplitude of message signal
fm=1000 # frequency of message signal
Ac=1 # amplitude of carrier signal
fc=10000 # frquency of carrier signal
Ka=1 # amplitude sensitivity


## ---- time-axis
Tstart=0
Tstop=4/fm
Tstep=0.0001/fm

t=np.arange(Tstart,Tstop,Tstep) # time
mt=Am*np.cos(2*math.pi*fm*t) # message
ct=Ac*np.cos(2*math.pi*fc*t) # carrier
st=(1+Ka*mt)*ct # normal AM modulation
At=np.abs(Ac*(1+Ka*mt))

f = plt.figure(figsize=(20,10))
f.subplots_adjust(hspace=0.5)
ax1 = f.add_subplot(311)
ax2 = f.add_subplot(312)
ax3 = f.add_subplot(313)

# plot the signals
ax1.plot(t,mt,'b',label='$m(t)$')
ax1.set_xlabel('time (sec)')
ax1.set_ylabel('$m(t)$ (v)')
ax1.set_title('Message Signal')
ax1.legend()
ax1.grid()

ax2.plot(t,ct,'k',label='$c(t)$')
ax2.set_xlabel('time (sec)')
ax2.set_ylabel('$c(t)$ (v)')
ax2.set_title('Carrier Signal')
ax2.legend()
ax2.grid()

ax3.plot(t,At,'b--',label='$+|A(t)|$')
ax3.plot(t,-1*At,'k--', label='$-|A(t)|$')
ax3.plot(t,st,'r',label='$s(t)$')
ax3.set_xlabel('time (sec)')
ax3.set_title('Modulated Signal')
ax3.set_ylabel('$s(t)$ (v)')
ax3.legend()
ax3.grid()
```
